```python
import pandas as pd 
import numpy as np
```

```python
df_omicron=pd.read_csv('omicron.csv')
```

```python
df_omicron
```

```python
df_india=df_omicron[df_omicron['Country']=='India']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#plt.boxplot(df_india['Date'],df_india['Percentage of Omicron'])
fig, axes = plt.subplots(figsize=(14, 8))
plt.title("Percentage of Omicron in Countries")
sns.lineplot(x='Date', y='Percentage of Omicron', hue='Country', data=df_omicron, ax=axes)
```

```python
fig, axes = plt.subplots(figsize=(14, 8))
plt.title('Omicron trends in india')
sns.lineplot(x='Date', y='Percentage of Omicron', data=df_india, ax=axes)
```

```python
five_country=df_omicron[(df_omicron["Country"]=="India") | (df_omicron["Country"]=="Germany") | (df_omicron["Country"]=="Canada") | (df_omicron["Country"]=="Japan") | (df_omicron["Country"]=="Russia")]
```

```python
fig, axes = plt.subplots(figsize=(14, 8))
plt.title("Percentage of Omicron in India, Germany, Canada, Japan, Russia")
sns.lineplot(x='Date', y='Percentage of Omicron', hue='Country', data=five_country, ax=axes)
```
